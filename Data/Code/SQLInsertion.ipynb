```python
import openpyxl
import csv
f = open('customer.csv', 'r', errors='ignore')
rdr = csv.reader(f);
member=[]
i=0
for line in rdr :
    i+=1
    member.append(line)
    print(line)
#member.append(line)
f.close()


ff = open('customer.txt','w');
for line in member:
    qquery= "insert into customer('customer_id', 'sex_id', 'age_id', 'nickname') values("+str(line[0])+", "+str(line[1])+", "+str(line[2])+", \""+line[3]+"\"  );"
    print(qquery);
    ff.write(qquery);
    ff.write("\n");
  

ff.close()
```
