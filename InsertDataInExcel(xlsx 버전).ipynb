```python
import pandas as pd
import numpy as np

import openpyxl
```

```python
filename = "stability.xlsx"
book = openpyxl.load_workbook(filename)
sheet = book.worksheets[0]
```

```python
data = []
```

```python
for row in sheet.rows :
    data.append([
        row[0].value,
        row[1].value,
        row[2].value,
        row[3].value,
        row[4].value,
        row[5].value
    ])
```

```python
data
```

```python
f = open('stability.txt', 'w');
for line in data :
    qquery = "insert stability('company_id', 'year', 'capital', 'debt', 'dept_ratio', 'equity_capital_ratio', 'violation_of_the_law') values("+str(line[0])+", "+str(line[1])+", "+str(line[2])+", "+str(line[3])+", "+str(line[4])+", "+str(line[5])+" ); "
    f.write(qquery);
    f.write("\n")
    
    
f.close()    
```
